## 🕒 2. Working with `datetime` and `time`

```python
from datetime import datetime
import time

# Starting date and time
start = datetime.now()
print("Current time:", start.strftime("%Y-%m-%d %H:%M:%S"))

# Sleep for 2 seconds
print("Waiting for 2 seconds...")
time.sleep(2)
print("Done waiting!")

# Ending date and time
end = datetime.now()
print("Current time:", end.strftime("%Y-%m-%d %H:%M:%S"))
```

## 🧱 5. Writing and Importing Your Own Module

A Python module is simply a .py file that contains Python code — such as functions, classes, or variables — which you can import and use in other programs to keep your code organised and reusable.

People write their own modules in Python to:
* Organise code into reusable, logical chunks
* Avoid repetition by putting common functions in one place
* Make code easier to read, maintain, and test
* Share code across multiple projects or with other developers
* Encapsulate functionality, so they can focus on using it, not rewriting it

It’s a way to keep things clean, efficient, and scalable — especially as projects grow.

```python

# Write a simple module to file
with open("my_module.py", "w") as f:
    f.write("""
def greet(name):
    return f"Hello, {name}!"
""")

# Import and use it
import my_module
print(my_module.greet("student"))
```

This code creates a simple Python module (my_module.py) that defines a greet() function, then imports that module and calls the function to print a personalised greeting. It shows how to write, save, and reuse your own code in another file.

## 🛠️ 6. Mini Project: Create a Command-Line Style Utility

```python

import random
from datetime import datetime
from pathlib import Path

# Step 1: Create output folder
Path("logs").mkdir(exist_ok=True)

# Step 2: Greeting
name = input("Enter your name: ")

# Step 3: Pick a random quote
quotes = [
    "Keep pushing forward!",
    "Believe in yourself.",
    "Every expert was once a beginner.",
    "You are capable of amazing things."
]
quote = random.choice(quotes)

# Step 4: Create message with time
now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
message = f"[{now}] Hello, {name}. Here’s your quote: {quote}"

# Step 5: Save to file
with open("logs/quote_log.txt", "a") as f:
    f.write(message + "\n")

print("\n" + message)
print("✅ Quote saved to logs/quote_log.txt")
```
