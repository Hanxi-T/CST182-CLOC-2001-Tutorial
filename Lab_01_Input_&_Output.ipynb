# **Question #1** - Time Converter
---

You are required to create a program that takes in the input of the number of seconds, and outputs the equivalent number of minutes, hours, days.<br/><br/>


**<u>Example Run:</u>**<br/>
Time (s)?: **86400**<br/>
Minutes = 3600<br/>
Hours = 24<br/>
Days = 1<br/>


### Your Answer:

```python
# Student Name: Hanxi Teng
# Student Number: 1036124

# Add answer below this line
time_s = float(input("Time (s)?: "))

minutes = int(time_s / 60)
hours = int(time_s / (60 ** 2))
days = int(hours / 24)

print('Minutes =', minutes)
print('Hours =', hours)
print('Days =', days)
```

# **Question #2** - Tickets Question
---

You've been hired to create a program to calculate the cost of tickets for events.

The ticket itself is \$20.00, but there is a booking fee of \$5.00 and a convience fee of \$7.00 per ticket. Once this subtotal is complete, a tax of 10% is added and then on top of this is a service charge of 5%.

The program needs to ask the user how many tickets they wish to purchase, and then the program needs to calculate and display all the above charges, and then lastly, the total sum owed by the customer.<br/><br/>

**<u>Example Run:</u>**<br/>
This program is meant to provide you with experience if understanding of a problem statement without directly showing the input/outputs - so no example run has been provided.

### Your Answer:

```python
# Student Name: Hanxi Teng
# Student Number: 1036124

# Add answer below this line
num_ticket = int(input("Enter the number of tickets you want to purchase: "))
ticket = 20
booking_fee = 5
convenience_fee = 7

ticket_cost = ticket * num_ticket
booking_fee_total = booking_fee
convenience_fee_total = convenience_fee * num_ticket


subtotal = ticket_cost + booking_fee_total + convenience_fee_total
tax = subtotal * 0.1

total_after_tax = subtotal + tax
service_charge = total_after_tax * 0.05

total_cost = ticket_cost + booking_fee_total + convenience_fee_total + tax + service_charge


# Display the results
print(f"Ticket cost: ${ticket_cost:.2f}")
print(f"Booking fee: ${booking_fee_total:.2f}")
print(f"Convenience fee: ${convenience_fee_total:.2f}")
print(f"Subtotal: ${subtotal:.2f}")
print(f"Tax (10%): ${tax:.2f}")
print(f"Service charge (5%): ${service_charge:.2f}")
print(f"Total amount owed: ${total_cost:.2f}")
```
