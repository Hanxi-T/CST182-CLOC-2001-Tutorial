```python
# Reference: https://docs.python.org/3/library/json.html
# To save and load the data
import json

# Make a list to store all the finance data
data = []

# Read data from the file
try:
    file = open("personal_finance_tracker.json", "r")
    data = json.load(file)
    file.close()
except:
    data = []

# Start a loop to keep the program running
while True:

# Give a name for the finance tracker
    print("")
    print("Your Best Personal Finance Tracker")
    print("")

# Show the Options List
    print("[1] Record Income")
    print("[2] Add Expense")
    print("[3] View Expenses by Category")
    print("[4] See the Amount You Saved")
    print("[5] Show the Finance Summary")
    print("[0] Save & Exit")
    print("")

    option = input("Enter Your Option: ")



# Choose Option [1] to record income
    if option == "1":

        print("")
        print("!! You Want to Record Income !!")
        print("")

        income_record = input("Enter Your Income ($): ")
        income = float(income_record)

        source_income = input("Enter the Source of Income: ")

# Create an empty dictionary to store one transaction
        record = {}
        record["type"] = "income"
        record["amount"] = income
        record["category"] = source_income

# Put the transaction into the list
        data.append(record)

        print("")
        print("Your Income Is Recorded ($):", income)
        print("")



# Choose Option [2] to record expense
    elif option == "2":

        print("")
        print("!! You Want to Record Expense !!")
        print("")

        expense_record = input("Enter Your Expense ($): ")
        expense = float(expense_record)

        category_expense = input("Enter the Category of Expense: ")

# Create an empty dictionary to store one transaction
        record = {}
        record["type"] = "expense"
        record["amount"] = expense
        record["category"] = category_expense

# Put the transaction into the list
        data.append(record)

        print("")
        print("Your Expense Is Recorded ($):", expense)
        print("")



# Choose Option [3] to view the expenses by category
    elif option == "3":

        print("")
        print("!! You Want to View the Expenses by Category !!")
        print("")

        category = input("Enter Category to View: ")

        total_expense = 0

        for i in data:

            if i["type"] == "expense":

                if i["category"] == category:

                  print("")
                  print("Amount ($):", i["amount"])
                  print("Category:", i["category"])
                  print("")

                  total_expense = total_expense + i["amount"]

        print("")
        print("Total Expense in the Category ($):", total_expense)
        print("")



# Choose Option [4] to see the amount you saved
    elif option == "4":

        print("")
        print("!! You Want to See the Amount You Saved !!")
        print("")

        total_income = 0
        total_expense = 0

        for i in data:

            if i["type"] == "income":

                total_income = total_income + i["amount"]

            if i["type"] == "expense":

                total_expense = total_expense + i["amount"]

        saved_amount = total_income - total_expense

        print("")
        print("You saved ($):", saved_amount)
        print("")



# Choose Option [5] to see the finance summary
    elif option == "5":

        print("")
        print("!! You Want to See the Finance Summary !!")
        print("")

        total_income = 0
        total_expense = 0

        for i in data:

            if i["type"] == "income":

                total_income = total_income + i["amount"]

            if i["type"] == "expense":

                total_expense = total_expense + i["amount"]

        balance = total_income - total_expense

        print("")
        print("Finance Summary")
        print("Your Total Income ($):", total_income)
        print("Your Total Expense ($):", total_expense)
        print("Balance ($):", balance)
        print("")



# Choose Option [0] to save the data and exit
    elif option == "0":

      print("")
      print("!! Data is Saving...... !!")
      print("")

# Write data to a file
      file = open("personal_finance_tracker.json", "w")
      json.dump(data, file)
      file.close()

      print("")
      print("!! Data Saved !!")
      print("")

# End the loop
      break

    else:
         print("")
         print("Invalid Option !! Please Choose From 0, 1, 2, 3, 4 or 5 !!")
         print("")
```
